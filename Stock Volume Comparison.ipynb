```python
from pandas_datareader import data as pdr
import matplotlib.pyplot as plt
import datetime

sams = pdr.get_data_yahoo('005930.KS', '2017-01-01')

plt.figure(figsize=(9,6))
plt.subplot(2,1,1)
plt.title('Samsung Electronics (Yahoo Finance)')
plt.plot(sams.index, sams['Close'] / 1100, 'c', label='Close') # KRW to USD : $1: \1100
plt.plot(sams.index, sams['Adj Close'] / 1100 , 'b--', label='Adj Close')
plt.legend(loc='best')
plt.subplot(2,1,2)
plt.bar(sams.index, sams['Volume'], color='g', label='Volume')
plt.legend(loc='best')
plt.show()
```

```python
app = pdr.get_data_yahoo('AAPL', start='2017-01-01')

plt.figure(figsize=(9,6))
plt.subplot(2,1,1)
plt.title('APPLE (Yahoo Finance)')
plt.plot(app.index, app['Close'], 'c', label='Close')
plt.plot(app.index, app['Adj Close'], 'b--', label='Adj Close')
plt.legend(loc='best')
plt.subplot(2,1,2)
plt.bar(app.index, app['Volume'], color='g', label='Volume')
plt.legend(loc='best')
plt.show()
```
